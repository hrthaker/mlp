```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from matplotlib import rcParams

%matplotlib inline

img = cv2.cvtColor(cv2.imread('img1.jpeg'),cv2.COLOR_BGR2RGB)
r,c=img.shape[:2]
kernel_identity = np.array([[0,0,0],[0,1,0],[0,0,0]])
kernel_3x3 = np.ones((3,3),np.float32)/9.0
kernel_5x5 = np.ones((5,5),np.float32)/25.0
k1 = cv2.filter2D(img,-1,kernel_identity)
k2 = cv2.filter2D(img,-1,kernel_3x3)
k3 = cv2.filter2D(img,-1,kernel_5x5)
rcParams['figure.figsize'] = 15 ,8
fig,ax = plt.subplots(1,4)
ax[0].imshow(img)
ax[1].imshow(k1)
ax[2].imshow(k2)
ax[3].imshow(k3)
plt.imshow(img)
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('sudoku1.jpg',0)

laplacian = cv2.Laplacian(img,cv2.CV_64F)
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)
canny = cv2.Canny(img,50,240)

plt.subplot(2,3,1),plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(2,3,2),plt.imshow(laplacian,cmap = 'gray')
plt.title('Laplacian'), plt.xticks([]), plt.yticks([])
plt.subplot(2,3,3),plt.imshow(sobelx,cmap = 'gray')
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])
plt.subplot(2,3,4),plt.imshow(sobely,cmap = 'gray')
plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])
plt.subplot(2,3,5),plt.imshow(canny,cmap = 'gray')
plt.title('Canny'), plt.xticks([]), plt.yticks([])

plt.show()
```
